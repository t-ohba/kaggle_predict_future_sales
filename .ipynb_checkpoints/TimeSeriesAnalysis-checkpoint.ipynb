### データ外観

```python
ts = train_df.groupby(['date_block_num'])['item_cnt_day'].sum()
```

```python
plt.plot(ts)
plt.xlabel('date_block_num')
plt.ylabel('sum of item_cnt_day')
plt.grid(True)
plt.show()
```

#### 季節性
- Multiplicative model: Y[t] = Trend[t] * Seasonality[t] * error[t]
- Additive model: Y[t] = Trend[t] + Seasonality[t] + error[t]

```python
res = sm.tsa.seasonal_decompose(ts.values, freq=12, model='multiplicative')
fig = res.plot()
fig.show()
```

```python
res = sm.tsa.seasonal_decompose(ts.values, freq=12, model='additive')
fig = res.plot()
fig.show()
```

#### 定常性
- 定常
    - 時系列データの平均が時間変化しない
    - 時系列データの分散が時間変化しない（均一分散性：homoscedasticity）
    - 時系列データの自己共分散が時間変化しない
- ADF(Augmented Dickley-Fuller)検定: 時系列データが単位根過程（非定常過程）かを調べる
    - p-valueが小さければ、帰無仮説「時系列データが非定常過程」を棄却

```python
def test_stationarity(time_series):
    adf_test = adfuller(time_series, autolag='AIC')
    print('Results of Augmented Dickey-Fuller Test:')
    output = pd.Series(adf_test[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observation Used'])
    for key, value in adf_test[4].items():
        output['Critical Value({})'.format(key)] = value
    print(output)
```

```python
test_stationarity(ts)
```

```python
def difference(data, interval=1): # i時点前との差分
    diff = list()
    for i in range(interval, len(data)):
        value = data[i] - data[i - interval]
        diff.append(value)
    return pd.Series(diff)
```

```python
plt.figure(figsize=(16, 16))
plt.subplot(3,1,1)
plt.plot(ts)
plt.title('original')
plt.xlabel('date_block_num')
plt.ylabel('sum of item_cnt_day')
plt.grid(True)

new_ts = difference(ts, 1) # トレンド成分の除去
plt.subplot(3,1,2)
plt.plot(new_ts)
plt.title('remove trend component')
plt.xlabel('date_block_num')
plt.ylabel('sum of item_cnt_day')
plt.grid(True)

new_ts = difference(ts, 12) # 季節成分の除去
plt.subplot(3,1,3)
plt.plot(new_ts)
plt.title('remove seasonal component')
plt.xlabel('date_block_num')
plt.ylabel('sum of item_cnt_day')
plt.grid(True)

plt.show()
```

```python
test_stationarity(difference(ts, 1))
```

```python
test_stationarity(difference(ts, 12))
```

### SARIMAモデル（季節自己回帰和分移動平均）

```python
new_ts = difference(ts, 1)
arma_order_select_ic(new_ts, ic='aic', trend='nc')
```

```python
sarimax = SARIMAX(ts,
                 order=(0, 1, 1), # order=(p, d, q)
                 seasonal_order=(0, 0, 0, 12), # seasonal_order(P, D, Q)[S]
                 enforce_stationarity=False,
                 enforce_invertibility = False).fit()
```

```python
sarimax.summary()
```

```python
plt.plot(ts, label='original')
plt.plot(sarimax.predict(0, 34), label='SARIMA')
plt.xlabel('date_block_num')
plt.ylabel('sum of item_cnt_day')
plt.legend()
plt.grid(True)
plt.show()
```

### Prophet

```python
train_df['date'] = train_df['date'].map(lambda x: datetime.datetime.strptime(x, '%d.%m.%Y'))
```

```python
ts = train_df.groupby('date_block_num')[['date', 'item_cnt_day']].agg({'date': 'min', 'item_cnt_day': 'sum'})
ts.columns=['ds', 'y']
model = Prophet(yearly_seasonality=True)
model.fit(ts)
```

```python
future = model.make_future_dataframe(periods=5, freq='MS')
forecast = model.predict(future)
forecast.tail()
```

```python
model.plot(forecast)
plt.show()
```

```python
model.plot_components(forecast)
plt.show()
```

```python
monthly_shop_sales = train_df.groupby([ 'date_block_num', 'shop_id'])['item_cnt_day'].sum()
monthly_shop_sales = monthly_shop_sales.unstack().fillna(0)
monthly_shop_sales['month'] = pd.date_range(start='2013-01-01', end='2015-10-01', freq='MS')
monthly_shop_sales.head()
```

```python
start = time.time()

result = {}
for shop_id in range(60):
    ts = pd.concat([monthly_shop_sales['month'], monthly_shop_sales.iloc[:, shop_id]], axis=1)
    ts.columns = ['ds', 'y']

    model = Prophet(yearly_seasonality=True)
    model.fit(ts)
    future = model.make_future_dataframe(periods=1, freq='MS')
    prediction = model.predict(future)
    result[shop_id] = prediction.tail(1)['yhat'][34]
    
elapsed_time = time.time() - start
print('elapsed time: {:.2f}[s]'.format(elapsed_time))
```
